```python
job_batch = {} # batch time of each job
for job, model in job_model.items():
    with open(f'{path}/7g.40gb_{model}.json') as f:
        read = json.load(f)
    time = []
    for key, val in read.items():
        time += val
    time = time[1:]
    job_batch[job] = np.mean(time)
job_batch
```
